Create a python class called WeatherStation with the following;
1. Attributes : 
        - name - name of the station
        - location - city/country

2. Method: 
        - report_weather(temp) - returns a sentence like: "station nairobi reports 22^c"

3. create an object from the class and call the method

```python
class WeatherStation ():
    def __init__ (self, name, location):
        self.name = name 
        self.location = location 

    def report_weather(self, temp):
        return f"Station {self.name} at {self.location} reports a temperature of {temp}°C."
    
WeatherStation1 = WeatherStation("Central Park", "New York")
print(WeatherStation1.report_weather(22))
```
